```python
import cv2
import numpy as np
import pandas as pd

def extract_pedestrian_blobs(video_path, output_dir,output_dir_1, model):
    # Load the video
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    bounding_boxes = pd.DataFrame(columns=['xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'class', 'name'])  # Create an empty DataFrame

    # Process each frame in the video
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Perform object detection
        results = model(frame)
        results = results.pandas()
        results = results.xyxy[0]
        # Filter pedestrian detections
        pedestrian_detections = results[results['class'] == 0]  # Assuming pedestrian class ID is 0

        # Select the nearest pedestrian
        if len(pedestrian_detections) > 0:
            pedestrian_detections['distance'] = np.abs((pedestrian_detections['xmin'] + pedestrian_detections['xmax']) / 2 - frame.shape[1] / 2)
            nearest_detection = pedestrian_detections.loc[pedestrian_detections['distance'].idxmin()]

            # Extract bounding box information
            x, y, x2, y2 = nearest_detection['xmin'], nearest_detection['ymin'], nearest_detection['xmax'], nearest_detection['ymax']
            confidence = nearest_detection['confidence']
            blob_filename = f"pedestrian_blob_{frame_count}.jpg"

            # Draw bounding box
            cv2.rectangle(frame, (int(x), int(y)), (int(x2), int(y2)), (0, 255, 0), 2)

            # Extract pedestrian blob
            pedestrian_blob = frame[int(y):int(y2), int(x):int(x2)]
            cv2.imwrite(f"{output_dir}/{blob_filename}", pedestrian_blob)

            # Store bounding box in the DataFrame
            row = pd.DataFrame([[x, y, x2, y2, confidence, 0, blob_filename]], columns=['xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'class', 'name'])
            bounding_boxes = bounding_boxes.append(row, ignore_index=True)

            # Save the modified frame
            modified_frame_filename = f"modified_frame_{frame_count}.jpg"
            cv2.imwrite(f"{output_dir_1}/{modified_frame_filename}", frame)
        frame_count += 1

        # Press 'q' to exit
        if cv2.waitKey(1) == ord('q'):
            break

    # Release resources
    cap.release()
    cv2.destroyAllWindows()

    return bounding_boxes
```

```python
import cv2
import numpy as np

def compute_object_center(contour_coordinates, image_height, image_width):
    # Create an empty binary mask
    mask = np.zeros((image_height, image_width), dtype=np.uint8)

    # Draw the contour on the binary mask
    cv2.drawContours(mask, [contour_coordinates], 0, 255, -1)

    # Calculate the moments of the binary mask
    moments = cv2.moments(mask)

    # Calculate the centroid coordinates
    centroid_x = int(moments['m10'] / moments['m00'])
    centroid_y = int(moments['m01'] / moments['m00'])

    return centroid_x, centroid_y
```
